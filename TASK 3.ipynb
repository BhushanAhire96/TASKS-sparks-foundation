# Bivarite Analysis

For category

```python
plt.figure(figsize=[10,5],dpi=80)
df.groupby('Category')['Sales'].mean().plot(kind='bar')
df.groupby('Category')['Sales'].mean()
```

```python
plt.figure(figsize=[10,5],dpi=80)
df.groupby('Category')['Discount'].mean().plot(kind='bar')
df.groupby('Category')['Discount'].mean()
```

```python
plt.figure(figsize=[10,5],dpi=80)
df.groupby('Category')['Profit'].mean().plot(kind='bar')
df.groupby('Category')['Profit'].mean()
```

```python
plt.figure(figsize=[10,5],dpi=80)
df.groupby('Category')['Quantity'].mean().plot(kind='bar')
df.groupby('Category')['Quantity'].mean()
```

# for sub-categories
it displays sales by subcategories

```python
plt.figure(figsize=[10,5],dpi=100)
df.groupby('Sub-Category')['Sales'].mean().plot(kind='barh',color='violet')
df.groupby('Sub-Category')['Sales'].mean()
```

```python
plt.figure(figsize=[10,5],dpi=100)
df.groupby('Sub-Category')['Quantity'].mean().plot(kind='barh',color='violet')
df.groupby('Sub-Category')['Quantity'].mean()
```

```python
plt.figure(figsize=[10,5],dpi=100)
df.groupby('Sub-Category')['Profit'].mean().plot(kind='barh',color='violet')
df.groupby('Sub-Category')['Profit'].mean()
```

```python
plt.figure(figsize=[10,5],dpi=100)
df.groupby('Sub-Category')['Discount'].mean().plot(kind='barh',color='violet')
df.groupby('Sub-Category')['Discount'].mean()
```

# for Region
Here we see region wise sales 

```python
plt.figure(figsize=[10,5],dpi=100)
df.groupby('Region')['Sales'].mean().plot(kind='barh',color='black')
plt.xlabel('Sales')
df.groupby('Region')['Sales'].mean()
```

```python
plt.figure(figsize=[10,5],dpi=100)
df.groupby('Region')['Discount'].mean().plot(kind='barh',color='black')
plt.xlabel('Discount')
df.groupby('Region')['Discount'].mean()
```

```python
plt.figure(figsize=[10,5],dpi=100)
df.groupby('Region')['Profit'].mean().plot(kind='barh',color='black')
plt.xlabel('Profit')
df.groupby('Region')['Profit'].mean()
```
